```python
import numpy as np
import pandas as pd
import joblib as jb
```

```python
longitude = np.arange(-51.27, -51.05, 0.005)
latitude = np.arange(-30.26, -29.96, 0.005)

caminhao = [False] # [True, False]
moto = [False] # [True, False]
cars = [False] # [True, False]
transport = [False] # [True, False]
others = [False] # [True, False]

holiday = [True, False]

day_of_week = np.arange(0, 7, 1)

hour = np.arange(0, 24, 1)

tipo_acid = ["ABALROAMENTO",
             "ATROPELAMENTO",
             "CHOQUE",
             "COLISÃO",
             "OUTROS"]
```

```python
index = pd.MultiIndex.from_product(
        [latitude,
         longitude,         
         caminhao,
         moto,
         cars,
         transport,
         others,
         holiday,
         day_of_week,
         hour,
         tipo_acid],
         names=[
                "latitude",
                "longitude",
                "caminhao",
                "moto",
                "cars",
                "transport",
                "others",
                "holiday",
                "day_of_week",
                "hour",
                "tipo_acid"
         ])
```

```python
df_predict = pd.DataFrame(index=index).reset_index()
```

```python
df_predict_trans = pd.get_dummies(
                        df_predict,
                        columns=["day_of_week", "hour"],
                        prefix=["day", "hour"],
                        drop_first=True)

df_predict_trans = pd.get_dummies(
                        df_predict_trans,
                        columns=["tipo_acid"],
                        prefix=["type"],
                        drop_first=True)
```

```python
df_predict_trans.head().T
```

```python
# Loading the scaler
scaler_feridos = jb.load("scaler_feridos.pkl")

# Loading the model
model_feridos = jb.load("model_feridos.pkl")
```

```python
X_scale = scaler_feridos.transform(df_predict_trans)
```

```python
y_predict = model_feridos.predict_proba(X_scale)
```

```python
df_probs = pd.DataFrame(y_predict, columns=["Prob_Not", "Prob_Yes"])
```

```python
df_predict = pd.concat([df_predict, df_probs], axis=1)
```

```python
df_predict.to_csv("acc_predict.csv")
```
